```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/water_potability.csv')
print(df)
```

```python
df = df.dropna()

# Describe statistics
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Water Quality Parameters')
plt.show()
```

```python
if 'pH' in df.columns and 'WQI' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x='pH', y='WQI', data=df)
    plt.title('pH vs Water Quality Index')
    plt.xlabel('pH')
    plt.ylabel('WQI')
    plt.show()
```
